```
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv('/content/Iris.csv',index_col=0)
print(iris.shape)
print(iris.head())
```

```
iris.plot(x='SepalLengthCm',y='SepalWidthCm')
```

```
iris.plot(x='SepalLengthCm',y='SepalWidthCm',kind='scatter')
plt.xlabel('Sepal Length cm')
plt.ylabel('Sepal Width cm')
plt.show()
```

```
iris.plot(y='SepalLengthCm',kind='box')
plt.ylabel('Sepal Length cm')
plt.show()
```

```
iris.plot(y='SepalLengthCm',kind='hist')
plt.ylabel('Sepal Length cm')
plt.show()
```

```
iris.plot(y='SepalLengthCm',kind='hist',bins=30,range=(4,8))
plt.ylabel('Sepal Length cm')
plt.show()
```
